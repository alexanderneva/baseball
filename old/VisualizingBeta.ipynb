```python
import numpy as np
from scipy.stats import beta as beta_dist
import matplotlib.pyplot as plt

# kappa=1 vs kappa=20, same mu=0.25
mu = 0.25
for kappa in [5, 10, 15, 20, 30, 50]:
    a = mu * kappa
    b = (1 - mu) * kappa
    x = np.linspace(0, 1, 200)
    plt.plot(x, beta_dist.pdf(x, a, b), label=f'kappa={kappa}')
plt.legend()
```
